```python
import numpy as np
from matplotlib import pyplot as plt

import sys
sys.path.insert(0, '../neural_networks/')
from somtf import SOM as SOMTF
from sompy import SOM as SOMPY
```

```python
def tag_generate(maps):
    dic = {}
    for i in np.unique(maps, axis=0):
        dic[tuple(i)] = []
    for i in range(len(maps)):
        dic[tuple(maps[i])].append(i)
            
    return dic
```

```python
train_data = np.loadtxt("../dataset/atividade_8/atividade_8_treino.txt", skiprows=1)
test_data = np.loadtxt("../dataset/atividade_8/atividade_8_teste.txt", skiprows=1)
```

```python
som = SOMTF(4, 4, 3, 100, alpha=.5, sigma=1, decay_in_time=False)
som.train(train_data)
image_grid = som.get_centroids()

train_mapped = som.map_vects(train_data)
test_mapped = som.map_vects(test_data)

# PLOT
plt.imshow(image_grid)
tags = tag_generate(train_mapped)
for key in tags:
    plt.text(key[1], key[0], len(tags[key]), ha='center', va='center',
             bbox=dict(facecolor='white', alpha=0.5, lw=0))
plt.title('SOM - numbers of samples mapped')
plt.show()

plt.imshow(image_grid)
tags = tag_generate(test_mapped)
for key in tags:
    plt.text(key[1], key[0], str(tags[key]), ha='center', va='center',
             bbox=dict(facecolor='white', alpha=0.5, lw=0))
plt.title('SOM - Table 8 predictions')
plt.show()
```
