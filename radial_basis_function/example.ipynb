```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from keras.optimizers import RMSprop
from keras.callbacks import EarlyStopping, History
import matplotlib.pyplot as plt
%matplotlib inline 

import sys
sys.path.insert(0, '../neural_networks/')
from rbf_layer import RBFLayer, InitCentersRandom
```

```python
def load_data():

    data = np.loadtxt("../dataset/functions/tanh.txt")
    X = data[:, :-1]
    y = data[:, -1:]
    return X, y
```

```python
if __name__ == "__main__":

    X, y = load_data() 

    model = Sequential()
    rbflayer = RBFLayer(1,
                        initializer=InitCentersRandom(X), 
                        betas=2.0,
                        input_shape=(1,))
    model.add(rbflayer)
    model.add(Dense(1))

    model.compile(loss='mean_squared_error',
                  optimizer=RMSprop())

    model.fit(X, y,
              batch_size=50,
              epochs=2000,
              verbose=0)

    y_pred = model.predict(X)
```

```python
print(rbflayer.get_weights()[0])
print()
print(rbflayer.get_weights()[1])
```

```python
plt.rcParams['figure.figsize'] = [15, 10]

plt.plot(X, y_pred, label='y_pred')
plt.plot(X, y, label = 'y')
plt.plot([-3,3], [0,0], color='black')
plt.xlim([-3,3])
    
centers = rbflayer.get_weights()[0]
widths = rbflayer.get_weights()[1]
plt.scatter(centers, np.zeros(len(centers)), s=20*widths)

plt.title('tanh(x) - 1 neuron')
plt.legend()
plt.show()
```
