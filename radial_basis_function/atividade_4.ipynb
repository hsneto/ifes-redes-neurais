```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from keras.optimizers import RMSprop
from keras.callbacks import EarlyStopping, History
import matplotlib.pyplot as plt
%matplotlib inline 

import sys
sys.path.insert(0, '../neural_networks/')
from rbf_layer import RBFLayer, InitCentersRandom
```

```python
def load_data():

    data = np.loadtxt("../dataset/atividade_4/atividade_4.txt")
#     data = np.loadtxt("../dataset/functions/tanh.txt")
    X = data[:, :-1]
    y = data[:, -1:]
    return X, y
```

```python
model = {} # model[nb_neurons][i-th trained model][model,
                                                 # yhat,
                                                 # history]
```

```python
if __name__ == "__main__":

    X, y = load_data()
    
    n = 5
    model[n] = {}
    
    rbflayer = RBFLayer(n,
                        initializer=InitCentersRandom(X), 
                        betas=2.0,
                        input_shape=(3,))
    
    for i in range(3):
        model[n][i] = {}

        model[n][i]['model'] = Sequential()       
        model[n][i]['model'].add(rbflayer)
        model[n][i]['model'].add(Dense(1)) # Output layer

        model[n][i]['model'].compile(loss='mean_squared_error',
                                     optimizer=RMSprop())

        earlyStopping = EarlyStopping(monitor='loss', 
                                      patience=10, 
                                      verbose=0, 
                                      mode='min')
        
        model[n][i]['history'] = History()
    
        model[n][i]['model'].fit(X, y,
                                 batch_size=50,
                                 epochs=2000,
                                 verbose=0,
                                 callbacks=[earlyStopping, model[n][i]['history']])

        model[n][i]['yhat'] = model[n][i]['model'].predict(X)
```

```python
plt.plot(range(len(model[5][0]['history'].history['loss'])), model[5][0]['history'].history['loss'], label='1-th training')
plt.plot(range(len(model[5][1]['history'].history['loss'])), model[5][1]['history'].history['loss'], label='2-th training')
plt.plot(range(len(model[5][2]['history'].history['loss'])), model[5][2]['history'].history['loss'], label='3-th training')
plt.legend()
```
